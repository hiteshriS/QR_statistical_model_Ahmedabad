```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

# Load India's shapefile
india = gpd.read_file("D:/India_Boundary/India_Boundary/India_Boundary.shp")
```

```python
# Coordinates of Ahmedabad
ahmedabad_lat = 23.0225
ahmedabad_lon = 72.5714

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 10))

# Plot India's shapefile
india.plot(ax=ax, color='white', edgecolor='black')

# Plot Ahmedabad
ax.plot(ahmedabad_lon, ahmedabad_lat, 'ro', markersize=5, label='Ahmedabad')

# Set plot extent to focus on Gujarat region
ax.set_xlim(65, 75)
ax.set_ylim(17.5, 25.5)

# Define grid lines (restricted)
lat_lines = np.arange(17.5, 25.51, 0.25)
lon_lines = np.arange(65, 75.01, 0.25)

# Draw grid lines only within the focused box
for lat in lat_lines:
    ax.plot([65, 75], [lat, lat], color='black', linewidth=0.2, linestyle='--')
for lon in lon_lines:
    ax.plot([lon, lon], [17.5, 25.5], color='black', linewidth=0.2, linestyle='--')

# Add labels and title
ax.set_title("Ahmedabad on India Map with 0.25 Grids", fontsize=14)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.legend()

plt.grid(False)
plt.tight_layout()
plt.savefig("D:/RUVISION/india_ahmedabad_grid.png", dpi=300, bbox_inches='tight')
plt.show()
```
