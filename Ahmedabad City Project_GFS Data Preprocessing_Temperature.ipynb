```python
columns_prec = []

for i in ['v']:
    for time_steps in forecast_hr:
        for j in np.arange(23.5,22.25,-0.25):
            for k in np.arange(72.0,73.25,0.25):
                columns_prec.append(f'{i}_{j}_{k}_{time_steps:03d}')
```

```python
data_prec = pd.DataFrame(index = pd.date_range(start = '2015-01-15', end = '2025-04-30 18:00' , freq = '6h'), columns = columns_prec)
```

```python
data_prec
```

```python

counter = 0

for time_step in data_prec.index:
    year = time_step.year
    month = time_step.month
    day = time_step.day
    ref_time = time_step.hour

    date_temp = pd.date_range(
        start=time_step + timedelta(hours=15),
        end=time_step + timedelta(hours=84),
        freq='3h'
    )
    col_temp = np.arange(0, 25)
    data_temp = pd.DataFrame(index=date_temp, columns=col_temp)

    for time_lag in time_from_ref:
        filename = f'gfs.0p25.{year}{month:02d}{day:02d}{ref_time:02d}.f{time_lag:03d}.grib2.nc'
        file_path = f"{root_directory}/{filename}"

        time = pd.to_datetime(f'{year}-{month}-{day}') + timedelta(hours=int(ref_time) + int(time_lag))

        if os.path.exists(file_path):
            X = Dataset(file_path, 'r')
            rel_hum = X.variables['V_GRD_L100'][:]
            lats = X.variables['lat'][:]
            lons = X.variables['lon'][:]

            latli = np.argmin(np.abs(lats - latbounds[1]))
            latui = np.argmin(np.abs(lats - latbounds[0]))
            lonli = np.argmin(np.abs(lons - lonbounds[0]))
            lonui = np.argmin(np.abs(lons - lonbounds[1]))

            arr = np.ravel(rel_hum[:, latli:latui, lonli:lonui])
            data_temp.loc[time, data_temp.columns[0:25]] = arr
            X.close()
        else:
            data_temp.loc[time, data_temp.columns[0:25]] = np.nan

    data_prec.loc[time_step, data_prec.columns[0:600]] = np.ravel(data_temp)

    counter += 1
    if counter % 500 == 0:
        print(f'Loop {counter} Done!')
```

```python
import os
import tarfile
import zipfile

src_dir = "/home/subhojit/sanghita_14-06-2025/GFS/V850"
dst_dir = os.path.join(src_dir, "extracted_files")
os.makedirs(dst_dir, exist_ok=True)

# Step 1: Extract all tar files into extracted_files and delete them
for fname in os.listdir(src_dir):
    if fname.endswith(".tar"):
        tar_path = os.path.join(src_dir, fname)
        try:
            with tarfile.open(tar_path) as tar:
                tar.extractall(path=dst_dir)
            print(f"Extracted tar: {fname}")
            os.remove(tar_path)
            print(f"Deleted tar: {fname}")
        except Exception as e:
            print(f"Error extracting tar {fname}: {e}")

# Step 2: Extract all zip files found in extracted_files and delete them
for root, dirs, files in os.walk(dst_dir):
    for fname in files:
        if fname.endswith(".zip"):
            zip_path = os.path.join(root, fname)
            try:
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    zip_ref.extractall(dst_dir)
                print(f"Extracted zip: {fname}")
                os.remove(zip_path)
                print(f"Deleted zip: {fname}")
            except Exception as e:
                print(f"Error extracting zip {fname}: {e}")
```
