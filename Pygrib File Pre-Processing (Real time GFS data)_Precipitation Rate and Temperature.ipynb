```python
def find_key(dictionary, element):
    for key, value in dictionary.items():
        if element in value:
            return key
    return 'None'
```

```python
# Replace 'root_directory' with the actual path to the folder containing the GRIB2 files
root_directory = 'D:/Temperature/Pygrib_real time data'

# Regular expression pattern to match the filenames in the format "gfs.t00z.pgrb2.0p25.f021"
filename_pattern = r'gfs\.\d{8}\.t\d{2}z\.pgrb2\.0p25\.f\d{3}$'

# Dictionary to store the filenames grouped by directory
directory_names = {}

# Function to check if the filename matches the expected pattern
def is_grib2_file(filename):
    return re.match(filename_pattern, filename)

# Walk through the root directory and its subdirectories
for dirpath, dirnames, filenames in os.walk(root_directory):
    x = str(dirpath).split(os.sep)[-1]
    directory_names[x] = []

    # Filter the filenames based on the expected pattern and store them in the dictionary
    for filename in filenames:
        if is_grib2_file(filename):
            directory_names[x].append(filename)

# Print the filenames of the GRIB2 files in each directory
for directory, filenames in directory_names.items():
    print(f"Directory: {directory}")
    for filename in filenames:
        print(filename)
    print()  # Add an empty line to separate directories
```

```python
# Extract dates from the filenames
dates = set()
for filenames in directory_names.values():
    for filename in filenames:
        # Extract date from filename (assuming the date is in the format 'YYYYMMDD')
        date_match = re.search(r'\d{8}\.t(\d{2})', filename)
        if date_match:
            dates.add(date_match.group(0))

# Create a DataFrame with dates present in the directories
start_date = min(dates)
end_date = max(dates)
data_tmp = pd.DataFrame(index=pd.date_range(start=start_date, end=end_date, freq='24H'), columns = columns_tmp)
```

```python
import pygrib
import pandas as pd
import numpy as np
import os
from datetime import timedelta

counter = 0

for time_step in data_tmp.index:
    
    year = time_step.year
    month = time_step.month
    day = time_step.day
    ref_time = time_step.hour

    # Create 3-hourly time index for forecast horizon
    date_temp = pd.date_range(start=time_step + timedelta(hours=3), end=time_step + timedelta(hours=72), freq='3H')
    col_temp = np.arange(0, 25)  # For a 5x5 grid

    data_temp = pd.DataFrame(index=date_temp, columns=col_temp)

    for time_lag in time_from_ref:
        
        filename = f'gfs.{year}{month:02d}{day:02d}.t{ref_time:02d}z.pgrb2.0p25.f{time_lag:03d}'
        directory = find_key(directory_names, filename)  # Ensure this returns only a subfolder or relative path

        time = pd.to_datetime(f'{year}-{month}-{day}') + timedelta(hours=int(ref_time) + int(time_lag))

        if directory != 'None':
            file_path = os.path.join(root_directory, directory, filename)

            try:
                grbs = pygrib.open(file_path)
                grb = grbs.select(name='2 metre temperature')[0]  # First 2m temperature field

                data = grb.values - 273.15  # Convert Kelvin to Celsius
                lats, lons = grb.latlons()

                # Get bounding box indices
                latli = 2
                latui = 7
                lonli = 2
                lonui = 7

                grid = data[latli:latui, lonli:lonui][::-1]  # Flip for north-south orientation
                flat = np.ravel(grid)

                if len(flat) == 25:
                    data_temp.loc[time] = flat
                else:
                    print(f"Unexpected grid size at {filename}")
                    data_temp.loc[time] = np.nan

                grbs.close()
            except Exception as e:
                print(f"Error reading {filename}: {e}")
                data_temp.loc[time] = np.nan
        else:
            print(f"File not found for {filename}")
            data_temp.loc[time] = np.nan

    # Store 3-day forecast values in the main DataFrame
    try:
        data_tmp.loc[time_step, data_tmp.columns[0:600]] = np.ravel(data_temp)
    except Exception as e:
        print(f"Error flattening or assigning data at {time_step}: {e}")

    counter += 1
    if counter % 500 == 0:
        print(f'Loop {counter} Done!')
```
