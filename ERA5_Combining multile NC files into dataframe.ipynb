```python
# Load the NetCDF file
#nc_files = glob.glob('D:/Ruvision/ERA5/V850/*.nc')
nc_files = glob.glob('D:/Ruvision/ERA5/V850/2015-17.nc')

# Step 2: Open multiple datasets   one
ds = xr.open_mfdataset(nc_files, combine='by_coords', engine='netcdf4')

# See the structure of the dataset
print(ds)
```

```python
# 2. Prepare the empty DataFrame
columns = []
latitudes = np.arange(17.5,25.75,0.25)
longitudes = np.arange(65,75.25,0.25)

for lat in latitudes:
    for lon in longitudes:
        columns.append(f'MSLP_{lat}_{lon}')
```

```python
data = pd.DataFrame(index = pd.date_range(start = '2015-01-01', end = '2024-12-31 21:00' , freq = '3H'), columns = columns)
data
```

```python
total = len(latitudes) * len(longitudes)
count = 0

for lat in latitudes:
    for lon in longitudes:
        # Extract nearest point
        sel_tp = ds['msl'].sel(latitude=lat, longitude=lon, method='nearest')

        # Convert to Pandas Series
        tp_series = pd.Series(sel_tp.values.ravel(), index=pd.to_datetime(ds.valid_time.values))

        # Reindex to match DataFrame index
        tp_series = tp_series.reindex(data.index)

        # Add to DataFrame
        col_name = f'MSLP_{lat}_{lon}'
        data[col_name] = tp_series

        # Update counter and show progress
        count += 1
        percent = (count / total) * 100
        print(f"{count}/{total} completed ({percent:.2f}%)")
```

```python
data
```

```python
total_nulls = data.isnull().sum().sum()
print(f"Total null values: {total_nulls}")
```

```python
group_idx = (data.index.to_series().reset_index(drop=True).index // 8)
summed_data = data.groupby(group_idx).mean()
```
