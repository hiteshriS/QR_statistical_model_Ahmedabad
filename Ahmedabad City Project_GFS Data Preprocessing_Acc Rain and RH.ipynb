```python
def find_key(dictionary, element):
    for key, value in dictionary.items():
        if element in value:
            return key
    return 'None'
```

```python
directory_names = {}
```

```python
for dirpath, dirnames, filenames in os.walk(root_directory):
    if dirpath.lower().endswith(".nc"):
        
        x = str(dirpath.lower()).split('\\')[-1]
        directory_names[x] = []
        
        for filename in filenames:
            directory_names[x].append(filename)
```

```python
columns_prec = []

for i in ['Prec']:
    for time_steps in forecast_hr:
        for j in np.arange(22.5 - 0.25, 23.5):
            for k in np.arange(65,75.25,0.25):
                columns_prec.append(f'{i}_{j}_{k}_{time_steps:03d}')
```

```python
counter = 0
for time_step in data_prec.index:
    
    year = time_step.year
    month = time_step.month
    day = time_step.day
    ref_time = time_step.hour

    date_temp = pd.date_range(start=time_step + timedelta(hours=15),
                              end=time_step + timedelta(hours=84),
                              freq='3h')
    col_temp = np.arange(0, 25)
    data_temp = pd.DataFrame(index=date_temp, columns=col_temp)

    for time_lag in time_from_ref:

        filename = f'gfs.0p25.{year}{month:02d}{day:02d}{ref_time:02d}.f{time_lag:03d}.grib2.nc'
        directory = find_key(directory_names, filename)

        if directory != 'None':

            X = Dataset(f"D:/Acc_Prec/{directory}/{filename}", 'r')
            time_var = X.variables['time']
            time_bnds = X.variables['time_bnds']
            prec = X.variables['A_PCP_L1_Accum_1']

            lats = X.variables['lat'][:]
            lons = X.variables['lon'][:]

            latli = np.argmin(np.abs(lats - latbounds[1]))
            latui = np.argmin(np.abs(lats - latbounds[0]))
            lonli = np.argmin(np.abs(lons - lonbounds[0]))
            lonui = np.argmin(np.abs(lons - lonbounds[1]))

            time = pd.to_datetime(f'{year}-{month}-{day}') + timedelta(hours=int(ref_time) + int(time_lag))
            flat_prec = np.ravel(prec[:, latli:latui, lonli:lonui])

            if (((int(time_lag) % 6) == 0) and (time_bnds[0][1] % 6 == 0) and (time_bnds[0][1] % 3 == 0)):
                time_prev = time - timedelta(hours=3)

                if len(flat_prec) != 25:
                    if (time_bnds[1][1] - time_bnds[1][0]) == 6:
                        data_temp.loc[time, data_temp.columns[0:25]] = flat_prec[25:] - np.ravel(data_temp.loc[time_prev, data_temp.columns[0:25]])
                    else:
                        print(filename)

                elif (time_bnds.shape[0] == 1) and (len(flat_prec) == 25):
                    data_temp.loc[time, data_temp.columns[0:25]] = flat_prec - np.ravel(data_temp.loc[time_prev, data_temp.columns[0:25]])

            elif (((int(time_lag) % 6) != 0) and (time_bnds[0][1] % 6 != 0) and (time_bnds[0][1] % 3 == 0)):

                if len(flat_prec) != 25:
                    if (time_bnds[1][1] - time_bnds[1][0]) == 3:
                        data_temp.loc[time, data_temp.columns[0:25]] = flat_prec[25:]
                    else:
                        print(filename)

                elif (time_bnds.shape[0] == 1) and (len(flat_prec) == 25):
                    data_temp.loc[time, data_temp.columns[0:25]] = flat_prec

    data_prec.loc[time_step, data_prec.columns[0:600]] = np.ravel(data_temp)

    counter += 1
    if counter % 500 == 0:
        print(f'Loop {counter} Done!')
```
