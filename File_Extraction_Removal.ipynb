```python
import os

def remove_files_with_extension(folder_path, extensions_to_remove):
    for filename in os.listdir(folder_path):
        for extension in extensions_to_remove:
            if filename.endswith(extension):
                file_path = os.path.join(folder_path, filename)
                os.remove(file_path)
                print(f"Removed: {filename}")

# Replace 'folder_path' with the actual path to your folder.
folder_path = 'F:/GFS FORECASTS_PREC_FLUX/gfs.0p25.2021060818.f072-25.2021112312.f048.grib2.nc'

# List of extensions to remove.
extensions_to_remove = ['.f099.grib2.nc', '.f102.grib2.nc']

# Call the function to remove the specified files.
remove_files_with_extension(folder_path, extensions_to_remove)
```

```python
import os
import zipfile

folder_path = 'F:/NCEP GFS/New folder/Acc_Prec/gfs.0p25.2019100506.f036-25.2022011206.f060.grib2.nc'

def extract_nc_from_zip(folder_path):
    # Get a list of all files in the folder
    files = os.listdir(folder_path)
    
    # Iterate through the files in the folder
    for file in files:
        # Check if the file is a WinZip file (ZIP archive)
        if file.endswith('.zip'):
            zip_file_path = os.path.join(folder_path, file)
            
            # Extract the contents of the ZIP file into the same directory
            with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                zip_ref.extractall(folder_path)
                
            print(f"Extracted: {zip_file_path}")
            
# Call the function to extract .nc files into the same directory as the ZIP files
extract_nc_from_zip(folder_path)
```

```python
import os
folder_path = 'F:/NCEP GFS/New folder/Acc_Prec/gfs.0p25.2019100506.f036-25.2022011206.f060.grib2.nc'

def remove_winzip_files(folder_path):
    # Get a list of all files in the folder
    files = os.listdir(folder_path)
    
    # Iterate through the files in the folder
    for file in files:
        # Check if the file is a WinZip file (ZIP archive)
        if file.endswith('.zip'):
            file_path = os.path.join(folder_path, file)
            
            # Remove the WinZip file
            os.remove(file_path)
            print(f"Removed: {file}")
remove_winzip_files(folder_path)
```
